## Renombrar columnas:


*   Renombrar columna 0 a “Flow_ID”
*   Renombrar columna 1 a “Date”

*   Renombrar columna 3 a “Count”

```python
df = df.rename(columns={0: 'Flow_ID',
                        1: 'Date', 
                        2: 'Time',
                        3: 'Count'})

df.head()
```

## Seleccionar datos
 
*   Seleccionar todos los datos para la fecha 24 de julio de 2005 con un flow id=7
*Con dos datos seleccionados en el paso 1, ordenen el resultado en Count en orden descendiente y saquen las 10 primeras filas. Asignen la salida a una variable llamada top 10.

```python
#Aquí voy a cambiar el tipo de datos para la columna Date y Time

df['Date'] = df['Date'].astype(int)
df['Time'] = df['Time'].astype(int)
df.info()
```

```python
# aquí voy a filtrar cuando la columna Date tiene el valor de fecha: 07/24/05
date_filter= df['Date']=='07/24/05'

date_filter.head()
```

```python
# Aquí voy a filtrar cuando la columna Flow_ID = 7
flow_filter= df['Flow_ID']==7

df[flow_filter]
```

```python
# ahora voy a agregar ambos filtros al mismo tiempo

dateFlow_filter = df.loc[date_filter & flow_filter, :]

dateFlow_filter
```
